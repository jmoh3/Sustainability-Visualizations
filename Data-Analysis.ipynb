```python
# df_state_to_centers.to_csv('dataViz/state-to-centers.csv')
df_state_to_centers = pd.read_csv('dataViz/state-to-centers.csv')
```

```python
def find_state_long(state):
    return df_state_to_centers[df_state_to_centers['State'] == state]['Longitude'].values[0]

def find_state_lat(state):
    return df_state_to_centers[df_state_to_centers['State'] == state]['Latitude'].values[0]
```

```python
df_summary
```

```python
longs = []
lats = []
for i in range(len(df_summary)):
    longs.append(find_state_long(df_summary.loc[i]['State']))
    lats.append(find_state_lat(df_summary.loc[i]['State']))
```

```python
df_summary['Longitude'] = longs
df_summary['Latitude'] = lats
```

```python
df_summary.drop('Unnamed: 40', axis=1, inplace=True)
```

```python
years = [str(x + 1990) for x in range(0, 27)]
```

```python
for year in range(1, len(years)):
    df_summary['Percent_Change_' + years[year] + '_1990'] = (df_summary[years[year]] - df_summary['1990']) / df_summary['1990'] * 100
```

```python
df_summary.to_csv('dataViz/residential_cleaned.csv')
```
