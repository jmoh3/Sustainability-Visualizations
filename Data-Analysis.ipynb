```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
df_summary = pd.read_csv('dataOriginal/summary_2016.csv', skiprows=2, nrows=51)
df_summary.rename(columns={'Million metric tons of carbon dioxide': 'State'}, inplace=True)
```

```python
df_usa = gpd.read_file('dataOriginal/usa.shp')
```

```python
def find_centroid(df, row):
    bounds = df['geometry'][row].bounds
    return (bounds[0] + bounds[2]) / 2, (bounds[1] + bounds[3]) / 2
```

```python
names = []
longs = []
lats = []
for i in range(len(df_usa)):
    centroid = find_centroid(df_usa, i)
    longs.append(centroid[0])
    lats.append(centroid[1])
    names.append(df_usa['STATE_NAME'][i])
```

```python
df_state_to_centers = pd.DataFrame.from_dict({'State': names, 'Longitude': longs, 'Latitude': lats})
```

```python
df_state_to_centers.to_csv('dataViz/state-to-centers.csv')
```

```python
def find_state_long(state):
    return df_state_to_centers[df_state_to_centers['State'] == state]['Longitude'].values[0]

def find_state_lat(state):
    return df_state_to_centers[df_state_to_centers['State'] == state]['Latitude'].values[0]
```

```python
df_summary
```

```python
longs = []
lats = []
for i in range(len(df_summary)):
    longs.append(find_state_long(df_summary.loc[i]['State']))
    lats.append(find_state_lat(df_summary.loc[i]['State']))
```

```python
df_summary['Longitude'] = longs
df_summary['Latitude'] = lats
```

```python
df_summary.drop('Unnamed: 30', axis=1, inplace=True)
```

```python
df_summary.to_csv('dataViz/summary_cleaned.csv')
```
