```python
import pandas as pd
import numpy as np
```

```python
df_list = []

for i in range(1, 42):
    add_on_df = pd.read_csv('cites_trade/trade_db_' + str(i) + '.csv', index_col=None, header=0)
    df_list.append(add_on_df)
    
df = pd.concat(df_list, axis=0, ignore_index=True)
```

```python
df
```

```python
year_df = df.groupby(by=['Exporter', 'Year']).size().reset_index()
```

```python
year_df = year_df.pivot(index='Exporter',columns='Year')
```

```python
year_df
```

```python
year_df.reset_index()
```

```python
# year_df.columns # = ['_'.join(col).strip() for col in major_df.columns.values]
year_df = year_df.fillna(0)
```

```python
year_df.columns = [str(x[1]) for x in year_df.columns.values]
```

```python
year_df = year_df.reset_index()
```

```python
year_df = year_df.rename(columns={'Exporter' : 'iso2'})
```

```python
year_df.columns
```

```python
country_codes = pd.read_csv('country-codes.csv')

country_codes
```

```python
def get_iso_3(code):
    if len(country_codes[country_codes['Alpha-2 code'] == code]) == 0:
        return 'NONE'
    return country_codes[country_codes['Alpha-2 code'] == code].values[0][2]
```

```python
year_df['iso3'] = year_df['iso2'].apply(get_iso_3)
```
